```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senoir']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create a column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' :90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
    
```

```python
from dotenv import load_dotenv
import os
load_dotenv() # Loaad .env into environment

from sqlalchemy import create_engine
# MYSQL Connection 

schema ="customer_behaviour"
host = "127.0.0.1"
user ="root"
password = os.getenv("MYSQL_PASSWORD")
port =" 3306"

connection_string = f'mysql+pymysql://{user}:{password}@{host}:{port}/{schema}'
engine =create_engine(connection_string)

print("Connection Successful!")

#Write DataFrame to MySQL
table_name ="customer_shopping" 
df.to_sql(table_name, engine,if_exists="replace" , index=False)
# Read back sample
pd.read_sql("SELECT * FROM customer_shopping LIMIT 5;", engine)
          
```
